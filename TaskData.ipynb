```
df = pd.read_json (r'source_file_2.json')
```

```
df = df.explode('managers')
```

```
df = df.explode('watchers')
```

```
df
```

```
dd = df.groupby(['managers', 'priority'])
dd.head()
```

```
manager = df['managers'].unique()
watcher = df['watchers'].unique()
```

```
cols = ['name', 'project', 'priority']
```

```
managers = pd.DataFrame(columns=cols)
i = 0
for m in manager:
  proj = list(set(df.name[df['managers'] == m]))
  prt = list(set(df.priority[df['managers'] == m]))
  for it in range(len(prt)):
    managers.loc[i+it,'name'] = m
    managers.loc[i+it,'project'] = proj[it]
    managers.loc[i+it,'priority'] = prt[it]
  i = i+ len(prt)
managers
```

```
watchers = pd.DataFrame(columns=cols)
i = 0
for m in watcher:
  proj = list(set(df.name[df['watchers'] == m]))
  prt = list(set(df.priority[df['watchers'] == m]))
  for it in range(len(prt)):
    watchers.loc[i+it,'name'] = m
    watchers.loc[i+it,'project'] = proj[it]
    watchers.loc[i+it,'priority'] = prt[it]
  i = i+ len(prt)
watchers
```

```
wtch = {}
for w in watcher :
  nw = watchers.groupby('name').get_group(w)
  nw = nw.sort_values('priority')
  lst = nw['project'].to_list()
  wtch.update({w:lst})

wtch
```

```
mngr = {}
for w in manager :
  nw = managers.groupby('name').get_group(w)
  nw = nw.sort_values('priority')
  lst = nw['project'].to_list()
  mngr.update({w:lst})

mngr
```

```
with open("managers.json", "w") as fm:  
    json.dump(mngr, fm) 
with open("watchers.json", "w") as fw:  
    json.dump(wtch, fw) 
```
